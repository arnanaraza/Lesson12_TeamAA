```python
## Import necessary modules
from osgeo import gdal
from osgeo.gdalconst import GA_ReadOnly, GDT_Float32
import numpy as np
import os
import tarfile
import glob
import os
import subprocess
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# check working directory if needed:
os.getcwd()
```

```python
#Preliminaries - directory setup, acquiring files 

#checks if data folder exists, otherwise creates it at main directory level. 
if not os.path.exists("data"):
    os.mkdir("data")

#download file to data folder
subprocess.call(['wget','https://www.dropbox.com/s/zb7nrla6fqi1mq4/LC81980242014260-SC20150123044700.tar.gz?dl=1',"-O", "data/landsat8.tar.gz"]) 

#extract all files from data folder
tar = tarfile.open('data/landsat8.tar.gz')
tar.extractall('data')
tar.close()
```

```python
#sort the files for proper indexing of bands then assign band 4 and 5
path= 'data/'
allTifs = glob.glob(path + "*.tif")
allTifsSort = sorted(allTifs)  
band4tif = allTifsSort [4]
band5tif = allTifsSort [5]
```

```python
#Define function 
def CompNDWI(band4, band5):
    #get tifs
    band4 = gdal.Open(band4tif, GA_ReadOnly)
    band5 = gdal.Open(band5tif, GA_ReadOnly)
    #set geo-transform
    geotransform = band4.GetGeoTransform()
    geotransform = band5.GetGeoTransform()
    #read bands as array
    band4Arr = band4.ReadAsArray(0,0,band4.RasterXSize, band4.RasterYSize)
    band5Arr = band5.ReadAsArray(0,0,band5.RasterXSize, band5.RasterYSize)
    #set data type as float
    band4Arr=band4Arr.astype(np.float32)
    band5Arr=band5Arr.astype(np.float32)
    #make a mask of the two bands
    mask = np.greater(band4Arr+band5Arr,0)
    #make NDWI and avoid warning of NAs
    with np.errstate(invalid='ignore'):
        ndwi = np.choose(mask,(-99,(band4Arr-band5Arr)/(band4Arr+band5Arr)))
    #check the real minimum value
    print 'real minimum is:'
    print ndwi[ndwi>-99].min()
    #write to disk    
    driver = gdal.GetDriverByName('GTiff')
    outDataSet=driver.Create('data/ndwi.tif', band4.RasterXSize, band4.RasterYSize, 1, GDT_Float32)
    outBand = outDataSet.GetRasterBand(1)
    outBand.WriteArray(ndwi,0,0)
    outBand.SetNoDataValue(-99)
    #set the projection and extent information of the dataset
    outDataSet.SetProjection(band4.GetProjection())
    outDataSet.SetGeoTransform(band5.GetGeoTransform())
    #savinglike in the OGR example flush it
    outBand.FlushCache()
    outDataSet.FlushCache()

CompNDWI(band4tif, band5tif)
```

```python
# Visualize NDWI

#open data
dsll = gdal.Open("data/ndwi_wgs.tif")

#read raster data
ndwi = dsll.ReadAsArray(0, 0, dsll.RasterXSize, dsll.RasterYSize)

#now plot the raster data using gist_earth palette
plt.imshow(ndwi, interpolation='nearest', vmin=-1, vmax=1, cmap=plt.cm.gist_earth, clim=(0, 1.0))
plt.colorbar()
plt.show()
dsll = None
```
